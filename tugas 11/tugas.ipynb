```python
import numpy as np
import pandas as pd
from scipy.spatial.distance import cdist
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
```

```python
# Dataset skor kedisiplinan siswa
data = pd.DataFrame({
    "Siswa": [1, 2, 3, 4, 5],
    "dtw": [2, 3, 4, 1, 3],
    "dit": [4, 4, 3, 5, 2],
    "dmt": [4, 3, 2, 4, 1],
    "ddb": [3, 5, 5, 2, 3]
    })
```

```python
features = data[["dtw", "dit", "dmt", "ddb"]].values
```

```python
manhattan_dist = cdist(features, features, metric='cityblock')
```

```python
complete_linkage = linkage(manhattan_dist, method='complete')
average_linkage = linkage(manhattan_dist, method='average')
```

```python
plt.figure(figsize=(10, 5))
dendrogram(complete_linkage, labels=data['Siswa'].values)
plt.title('Dendrogram - Complete Linkage')
plt.xlabel('Siswa')
plt.ylabel('Jarak')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
dendrogram(average_linkage, labels=data['Siswa'].values)
plt.title('Dendrogram - Average Linkage')
plt.xlabel('Siswa')
plt.ylabel('Jarak')
plt.show()
```
